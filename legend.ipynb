```python
import matplotlib.pyplot as plt
import numpy as np

def generate_rgb_combinations():
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.set_title('RGB Combinations for (a, b, b)')
    
    # Generate all combinations of RGB where G and B are the same
    r = np.arange(0, 256)
    gb = np.arange(0, 256)
    R, GB = np.meshgrid(r, gb)
    R = R.flatten()
    G = GB.flatten()
    B = GB.flatten()
    
    # Normalize the values to [0, 1] for matplotlib
    R_norm = R / 255.0
    G_norm = G / 255.0
    B_norm = B / 255.0
    
    # Plot each point with its corresponding color
    ax.scatter(R, G, B, c=np.stack([R_norm, G_norm, B_norm], axis=1))
    
    ax.set_xlabel('Red')
    ax.set_ylabel('Green')
    ax.set_zlabel('Blue')
    plt.show()

generate_rgb_combinations()
```
